## **Frequency Tables**

### *Product-Gender*

```python
# Creating a cross-tabulation of Product and Gender
product_gender_crosstab = pd.crosstab(aerofit_df['Product'], aerofit_df['Gender'])
```

```python
print(product_gender_crosstab)
```

```python
# Calculating percentages
total_customers = len(aerofit_df)
```

```python
# Percentage of male customers purchasing a treadmill
male_customers_percentage = round((product_gender_crosstab['Male'].sum() / total_customers) * 100, 0)

print("Percentage of Male Customers Purchasing a Treadmill:", male_customers_percentage)
```

```python
# Percentage of female customers purchasing KP781
female_customers_kp781_percentage = round((product_gender_crosstab.loc['KP781', 'Female'] / total_customers) * 100,0)

print("Percentage of Female Customers Purchasing KP781:", female_customers_kp781_percentage)
```

```python
# Probability of a customer being female given the product is KP281
p_female_given_kp281 = product_gender_crosstab.loc['KP281', 'Female'] / product_gender_crosstab.loc['KP281'].sum()

print("Probability of a Customer Being Female Given the Product is KP281:", p_female_given_kp281)
```

### *Product-Age*

```python
# Defining the age range
age_range = (20, 30)
```

```python
# Filtering customers within the age range
customers_in_age_range = aerofit_df[(aerofit_df['Age'] >= age_range[0]) & (aerofit_df['Age'] <= age_range[1])]
```

```python
# Creating'AgeGroup' column based on age range
customers_in_age_range['AgeGroup'] = pd.cut(customers_in_age_range['Age'], bins=[0, 20, 30, 40, 50, float('inf')], labels=['<20', '20-30', '30-40', '40-50', '50+'])
```

```python
# Creating a cross-tabulation of Product and AgeGroup
product_age_crosstab = pd.crosstab(customers_in_age_range['Product'], customers_in_age_range['AgeGroup'])

print(product_age_crosstab)
```

```python
# Calculating the total number of customers in the age range
total_customers_in_age_range = len(customers_in_age_range)

print("Total Number of Customers in the Age Range:", total_customers_in_age_range)
```

```python
# Calculating the percentage of customers for each product within the age range
product_percentages = (product_age_crosstab / total_customers_in_age_range) * 100

print(product_percentages)
```

**Observation:**
1. Age, Education, Usage and Fitness columns have very few outliers.
2. On the otherhand, Income and Miles columns a lot of outliers.

## **Treatment of Outliers with IQR Method**

```python
aerofit_df3 = aerofit_df.copy()
```

### *Removal of Outliers from Income*

```python
Q3 = aerofit_df3['Income'].quantile(0.75)
Q1 = aerofit_df3['Income'].quantile(0.25)
IQR = Q3 - Q1
aerofit_df3 = aerofit_df3[(aerofit_df3['Income'] > Q1 - 1.5*IQR) & (aerofit_df3['Income'] < Q3 + 1.5*IQR)]

sns.boxplot(data=aerofit_df3, x="Income", orient="h")
plt.show()
```

### *Removal of Outleirs from Miles*

```python
# Outlier Treatment: Removing top 15 & bottom 15 of the Column Outlier values
Q3 = aerofit_df3['Miles'].quantile(0.75)
Q1 = aerofit_df3['Miles'].quantile(0.25)
IQR = Q3 - Q1
aerofit_df3 = aerofit_df3[(aerofit_df3['Miles'] > Q1 - 1.5*IQR) & (aerofit_df3['Miles'] < Q3 + 1.5*IQR)]

sns.boxplot(data=aerofit_df3, x="Miles", orient="h")
plt.show()
```
